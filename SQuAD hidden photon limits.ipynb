```python
%matplotlib inline
from numpy import*
from pylab import*
import matplotlib.pyplot as plt

#STIXGeneral

font = {'family' : 'STIXGeneral',
        'weight' : 'normal',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
data = loadtxt('SQuAD_2021.txt', delimiter='\t')

data_r2 = loadtxt('SQuAD_2021_run2.txt', delimiter='\t')
```

```python
plt.figure(figsize=(8, 5))
plt.xscale('log')
plt.yscale('log')

plt.fill_between(data[0], data[1], data[2], color = 'mediumseagreen', alpha = 0.45, label = "Run 1 Exclusion")

plt.fill_between(data_r2[0], data_r2[1], data_r2[2], color = 'darkmagenta', alpha = 0.45, label = "Run 2 Exclusion")

plt.xlabel("DM Mass (micro-eV)")
plt.ylabel("Dark photon mixing angle")
plt.legend(loc = (0.65, 0.1))

m_min = 1/sqrt(10) *1e-8 * 4.135665538538e-6
m_max = sqrt(10) *1e6 * 4.135665538538e-6
ylim(1e-17, 1e-3)
xlim(m_min, m_max)

plt.figure(figsize=(6, 4))
plt.xscale('linear')
plt.yscale('log')

plt.fill_between(data[0]/1e-6, data[1], data[2], color = 'mediumseagreen', alpha = 0.45, label = "Run 1 Exclusion")

plt.fill_between(data_r2[0]/1e-6, data_r2[1], data_r2[2], color = 'darkmagenta', alpha = 0.45, label = "Run 2 Exclusion")

plt.xlabel("DM Mass (micro-eV)")
plt.ylabel("Dark photon mixing angle")
plt.legend(loc = (0.15, 0.8))
xlim(24.858, 24.863)
```
